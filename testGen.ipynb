# 01 AUTO DEPOSIT TO SMART WALLET

```python
# !pip install pygetwindow pywinauto
#! IMPORTANT: the window of Metamask browser should be at 80% and at wide position.
import pygetwindow as gw
from pywinauto.application import Application
import time
import pyautogui
import numpy as np
import cv2
import random
def templatePosOnScreen(template="template.png"):
    """
    Find the position of an image on the screen using template matching.

    Parameters:
        template (str): The path of the image to search for. Default is "template.png".

    Returns:
        tuple: A tuple containing the x and y coordinates of the center of the image.
    """
    # Capture a screenshot of the screen
    screenshot = pyautogui.screenshot()

    # Convert the screenshot to a NumPy array
    image = np.array(screenshot)

    # Load the image of the button
    button_image = cv2.imread(template)

    # Perform template matching
    result = cv2.matchTemplate(image, button_image, cv2.TM_CCOEFF_NORMED)

    # Find the coordinates of the best match
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

    # Draw a rectangle around the button
    top_left = max_loc
    bottom_right = (
        top_left[0] + button_image.shape[1],
        top_left[1] + button_image.shape[0],
    )
    cv2.rectangle(image, top_left, bottom_right, (0, 255, 0), 2)

    # Find the center of the button
    cx = top_left[0] + button_image.shape[1] // 2
    cy = top_left[1] + button_image.shape[0] // 2

    return (cx, cy)

def moveToThenClick(position):
    """
    Move the mouse cursor to a position and then click.

    Parameters:
        position (tuple): A tuple containing the x and y coordinates of the position to move to.
    """
    pyautogui.moveTo(position)
    print(f"Moved to {position}")
    pyautogui.click()
    print(f"Clicked at {position}")

def moveToThenClickImage(imagePath="send.png",sleepTime=1):
    pos = templatePosOnScreen(imagePath)
    moveToThenClick(pos)
    time.sleep(sleepTime)    

stepDicts=[
    {"clickImage":"IMGtemplate/send80.png","sleepTime":1,},
    {"type":"0x940c0db1462b8e5918d41ebd1337d30265cb34b2","sleepTime":2,},
    {"clickImage":"IMGtemplate/matic80.png","sleepTime":2,},
    {"clickImage":"IMGtemplate/tusdc80.png","sleepTime":1,},
    {"transactionValue":10,"sleepTime":1,},
    {"clickImage":"IMGtemplate/next80.png","sleepTime":3,},
    {"clickImage":"IMGtemplate/confirm80.png","sleepTime":3,},
]

def batchTestMetamask(transactionTimes=2,transactionHigh=10.0,transactionLow=0.0,manual=False):
    
    # Generate 50 random float numbers between 0 and 100 (adjust range as needed)
    random_floats = np.random.uniform(low=transactionLow, high=transactionHigh, size=transactionTimes)
    
    if manual==True:
        # Generate values within the range (100, 1000000) with a step of 5000
        values = np.arange(5000, 50000, 100)

        # Randomly select 50 values from the generated range
        random_values = np.random.choice(values, transactionTimes, replace=False)
        random_floats=random_values
    
    print(f"Random floats generated: {random_floats}")
    for s,transactionValue in enumerate(random_floats):
        for i,stepDict in enumerate(stepDicts):
            print(f"Step {i}: {stepDict}")
            if "clickImage" in stepDict:
                moveToThenClickImage(stepDict["clickImage"],stepDict["sleepTime"])
            if "type" in stepDict:
                pyautogui.typewrite(stepDict["type"])
                time.sleep(stepDict["sleepTime"])
            if "transactionValue" in stepDict:
                value=str(random_floats[s])
                print(f"Transaction value: {value}")
                pyautogui.typewrite(value)
                time.sleep(stepDict["sleepTime"])
            if "scrolldown" in stepDict:
                pyautogui.scroll(stepDict["scrolldown"])
                time.sleep(stepDict["sleepTime"])
        sleepTime=random.randint(3, 7)
        print(f"Going to sleep (second): {sleepTime}")
        time.sleep(sleepTime)

#! INIT setup
# Give you time to switch to the active browser window
time.sleep(5)

# Get the active window (assuming it's your browser)
active_window = gw.getActiveWindow()
window_title = active_window.title

# Print the active window's title
print(f"Active window title: {window_title}")

# Use Pywinauto to control the browser
app = Application().connect(title=window_title)

# Bring the window to the foreground
app.top_window().set_focus()

#! MAIN program
RUN_TIMES=50
batchTestMetamask(transactionTimes=RUN_TIMES,transactionLow=0.001,transactionHigh=0.1000,manual=True)
```
